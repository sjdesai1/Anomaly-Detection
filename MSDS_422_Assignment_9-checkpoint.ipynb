```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #for plotting
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
import datetime
from keras.layers import Dense, Input, Conv2D, LSTM, MaxPool2D, Flatten
from keras.models import Model
from keras.callbacks import EarlyStopping

%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()
import io
x_train_test_asis = pd.read_csv(io.BytesIO(uploaded['MNIST_A9_train.csv']))
kaggle_submission_asis = pd.read_csv(io.BytesIO(uploaded['MNIST_A9_test.csv']))
```

```python
y_train_test = x_train_test_asis.label
x_train_test = x_train_test_asis.drop(["label"],1)
```

```python
x_train_test = x_train_test.values.astype('float32') # all pixel values
y_train_test = y_train_test.astype('int32') # only labels i.e targets digits
kaggle_test = kaggle_submission_asis.values.astype('float32') # all pixel values
```

```python
%matplotlib inline
# preview the images first
plt.figure(figsize=(12,10))
x, y = 10, 5
for i in range(50):  
    plt.subplot(y, x, i+1)
    plt.imshow(x_train_test[i].reshape((28,28)),interpolation='nearest')
plt.show()
```

```python
#Normalising The Data
x_train_test = x_train_test/255.0
kaggle_test = kaggle_test/255.0
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train_test, y_train_test, test_size=0.2, random_state=42)
```

```python
print("xtrain shape: ",x_train.shape, "| x_test shape: ", x_test.shape, "| y_train.shape: ", y_train.shape, "| y_test.shape: ", y_test.shape)
```

Reshape to match keras input expectation

```python
x_train = x_train.reshape(x_train.shape[0], 28, 28,1)
x_test = x_test.reshape(x_test.shape[0], 28, 28,1)
```

Data looks good to proceed to the next step.

Now we can compile the Neural network.

Trial 1: Encoder, Latent Layer and Decoder

```python
encoder1 = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=[28, 28]),
  tf.keras.layers.Dense(100, activation='selu'),
  tf.keras.layers.Dense(30, activation='selu')
])

## latent view
#latent1   = Dense(10, activation='sigmoid')(encoder1)

decoder1 = tf.keras.models.Sequential([
  tf.keras.layers.Dense(100, activation='selu', input_shape=[30]),
  tf.keras.layers.Dense(28*28, activation='sigmoid'),
  tf.keras.layers.Reshape([28,28])
])
```

```python
model1 = keras.models.Sequential([encoder1, decoder1])  #change this to Model 1
model1.compile(loss='binary_crossentropy', optimizer='adam') #, metrics = ['accuracy'])
```

```python
history1 = model1.fit(x_train, x_train, epochs=20, validation_data=(x_test, x_test))
```

```python
# Plot the learning curve
pd.DataFrame(history1.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.title('Learning curve\n Mean loss, accuracy for training and validation sets measured over each epoch')
plt.show()
```
