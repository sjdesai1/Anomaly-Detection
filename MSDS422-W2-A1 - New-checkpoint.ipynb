```python
#missing data
total = housing_train.isnull().sum().sort_values(ascending=False)
percent = (housing_train.isnull().sum()/housing_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# Combining Train and test data for Feature engineering - taking care of outliers and missing data 
trainindex = housing_train.shape[0]
testindex = housing_test.shape[0]
y_train = housing_train.SalePrice.values
housing_data = pd.concat((housing_train, housing_test)).reset_index(drop=True)
housing_data.drop(['SalePrice'], axis=1, inplace=True)
print("housing_data (including train and test data) is : {}".format(housing_data.shape))
```

```python
#Label Encoding some categorical variables that may contain information in their ordering set

from sklearn.preprocessing import LabelEncoder
cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 
        'KitchenQual', 'BsmtFinType1', 'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope', 
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 'YrSold', 'MoSold')
# process columns, apply LabelEncoder to categorical features
for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(housing_data[c].values)) 
    housing_data[c] = lbl.transform(list(housing_data[c].values))
```

```python
housing_data = pd.get_dummies(housing_data)
#filling NA's with the mean of the column:
housing_data = housing_data.fillna(housing_data.mean())


#creating matrices for sklearn:
X_train = housing_data[:housing_train.shape[0]]
X_test = housing_data[housing_train.shape[0]:]
y = housing_train.SalePrice
```

```python
# Importing sklearn libraries
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, Lasso, LassoCV, LassoLarsCV
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
```

```python
# validation function to return rmse based on input model 
def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y, scoring="neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
# Ridge Regression 
alphas = [0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75]
cv_ridge = [rmse_cv(Ridge(alpha = alpha)).mean() 
            for alpha in alphas]

cv_ridge = pd.Series(cv_ridge, index = alphas)
#cv_ridge #[2] #.min()

cv_ridge[cv_ridge == cv_ridge.min()]
```

```python
X_test
```

```python
ridge_reg = Ridge(alpha=10, solver="cholesky", normalize=False)
ridge_reg.fit(X_train, y)
ridge_preds = np.expm1(ridge_reg.predict(X_test))

ridge_predss = pd.Series(ridge_preds)

ridge_predss.to_csv('ridge_preds_msds422a2.csv')
ridge_predss
```

```python
# Lasso Regression 
housing_lasso = make_pipeline(RobustScaler(), LassoCV(alphas = [1, 0.1, 0.001, 0.0005]).fit(X_train, y))
rmse_cv(housing_lasso).min() 
```

```python
rmse_cv(housing_lasso) #min is for alpha = 0.0005
```

```python
lasso_reg = Lasso(alpha=0.0005).fit(X_train, y)
lasso_preds = np.expm1(lasso_reg.predict(X_test))

lasso_predss = pd.Series(lasso_preds)

lasso_predss.to_csv('lasso_preds_msds422a2.csv')
lasso_predss
```

```python
# Elastic Net Regression 
housing_ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9))
rmse_cv(housing_ENet).min()
```

```python
enet_reg = ElasticNet(alpha=0.0005, l1_ratio=.9).fit(X_train, y)
enet_preds = np.expm1(enet_reg.predict(X_test))

enet_predss = pd.Series(enet_preds)

enet_predss.to_csv('enet_msds422a2.csv')
enet_predss
```
